```python
import pandas as pd
import numpy as np
import tensorflow as tf
from joblib import load,dump
```

```python
df=pd.read_csv("/kaggle/input/test-insurance-abi/test_insurance.csv")
df.head()
```

```python
ann=tf.keras.models.load_model("/kaggle/input/test-insurance-abi-v2/kaggle/best_model.keras")
lr=load("/kaggle/input/test-insurance-abi-v2/kaggle/lr_encoder.pkl")
sc=load("/kaggle/input/test-insurance-abi-v2/kaggle/sc_scaler.pkl")
```

```python
cat_col=list(df.select_dtypes('object'))
for i in cat_col:
    df[i]=lr.fit_transform(df[i])
df.head()
```

```python
ind=df.iloc[:,0].values
x=df.iloc[:,1:].values
print(x.shape)
```

```python
x=sc.fit_transform(x)
```

```python
y_pred=ann.predict(x)
y_pred=y_pred.flatten()
ans_df = pd.DataFrame({
    'id': ind,
    'Response': y_pred
})
ans_df.to_csv("answer.csv", index=False)
```
